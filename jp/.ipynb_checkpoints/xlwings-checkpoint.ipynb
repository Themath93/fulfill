## ship_ready
- 출고품목확인

```python
def ship_ready():
    """
    출고 행 확인
    """
    sel_rng = wb.selection.api.SpecialCells(12).Address.split(",")
    
    range_list = []

    for rng in sel_rng:

        if ":" in rng:
            num_0 = rng.split(":")[0].split("$")[-1]
            num_1 = rng.split(":")[1].split("$")[-1]
            #범위선택이 한 컬럼이 아닌 여러 컬럼을 동시에 선택된 경우
            if num_0 == num_1 :
                range_list.append(num_0)
            else : 
                range_list.append(str(num_0) + ',' + str(num_1))
        else :
            range_list.append(rng.split("$")[-1])
            
    df = pd.DataFrame()

    for rng in range_list :
        if ',' in rng :
            num_0 = rng.split(',')[0]
            num_1 = rng.split(',')[1]
            df = pd.concat([df,pd.DataFrame(wb.sheets[wb.selection.sheet.name].range("A" + num_0 + ":" + "R"+ num_1).value)])
        else :
            df = pd.concat([df,pd.DataFrame([wb.sheets[wb.selection.sheet.name].range("A" + rng + ":" + "R"+ rng).value])])
            
    row_nm = ' '.join(range_list)
    
    out_sht = wb.sheets['Shipment information']
    out_sht.range("C1:C6").clear_contents()
    
    out_sht.range("C1").value = row_nm
    out_sht.range("C3").value = "=TODAY()+1"
    
    wb.app.macro("MyMsgBox").run()
    
    out_sht.range("C1:C6").select()
```
